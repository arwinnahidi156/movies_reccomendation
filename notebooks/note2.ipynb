```python
movies=pd.read_csv("../data/movies.csv")
credits=pd.read_csv("../data/credits.csv")
ratings=pd.read_csv("../data/ratings.csv")
```

```python
movies_filtered=movies[movies['vote_average']!=0].copy()
movies_filtered
```

```python
m=movies_filtered['vote_count'].quantile(0.9)
m
```

```python
C=movies_filtered['vote_average'].mean()
```

```python
def weighted_rating(df,m=m,C=C):
  R=df['vote_average']
  v=df['vote_count']
  wr=(v/(v+m))*R + (m/(m+v))*C
  return wr
  

```

```python

movies_filtered['weighted_rating']=movies_filtered.apply(weighted_rating,axis=1)
```

```python
def similar_movies(movie_title,nr_movies):
  
  idx=movies_filtered.loc[movies_filtered['title']==movie_title].index 
  scores=list(enumerate(similarity_matrix[idx][0]))
  scores=sorted(scores,key=lambda x:x[1],reverse=True)
  movie_indices=[tpl[0] for tpl in scores[1:nr_movies+1]]
  movie_titles=[movies_filtered.loc[idx,'title'] for idx in movie_indices]
  return movie_titles

  
```
